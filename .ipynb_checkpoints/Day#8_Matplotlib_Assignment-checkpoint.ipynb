```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
plt.plot([1,2,3,4,5])
plt.plot(pd.DataFrame([6,7,8,9]));
plt.legend(['list','DataFrame'])
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(np.arange(1,5), explode=[0.1,0.1,0.1,0.1]);
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = np.random.randn(150)
y = np.random.randn(150)
plt.scatter(x,y, alpha=0.4, s=50)
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.title("Ziyad",loc="left",c="red")
plt.show()
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot([1,3,4,5],c ='y')
plt.plot(pd.DataFrame([1,6,7,8]), c ='r')
plt.plot([3,6,9,10],c ='b')
plt.legend(["lineA","lineB","lineC"],loc='upper left')
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])
plt.suptitle("Ziyad", fontsize=50)
plt.subplot(3,2,1)
x = np.random.randn(150)
y = np.random.randn(150)
plt.scatter(x,y,marker='+',alpha=0.4,s=50,c='r')
plt.subplot(3,2,3)
plt.show
```
